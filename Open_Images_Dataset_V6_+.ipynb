```
# Importing the libraries
import pandas as pd
import os
```

```
# Reading the csv file

os.chdir('/content/drive/My Drive/Colab Notebooks/DatasetnAnnotations/csv_folder')
classes_data = pd.read_csv('class-descriptions-boxable.csv', header=None)
classes_data.head()
```

```
# Define required classes
classes = ['Ambulance', 'Car', 'Taxi']

# Get the strings corresponding to the class name and store it inside a list called as class_strings
class_strings = []

for i in classes:
  # Extract the first column's elements for matching classes 
  req_classes = classes_data.loc[classes_data[1] == i]
  string = req_classes.iloc[0][0]
  print(string)
  # Append the strings corresponding to the class name into the list
  class_strings.append(string)

print(class_strings)
```

```
# Get only required columns from annotations csv file and get rid of NOT required columns

annotation_data_training = pd.read_csv('train-annotations-bbox.csv', usecols=['ImageID', 'LabelName', 'XMin', 'XMax', 'YMin', 'YMax'])
annotation_data_testing = pd.read_csv('test-annotations-bbox.csv', usecols=['ImageID', 'LabelName', 'XMin', 'XMax', 'YMin', 'YMax'])

annotation_data_training.head()
```

```
# Get only records with matching classes (LabelName = class_strings)

filtered_class_data_train = annotation_data_training.loc[annotation_data_training['LabelName'].isin(class_strings)].copy()
filtered_class_data_train.head()
```

```
filtered_class_data_test = annotation_data_testing.loc[annotation_data_testing['LabelName'].isin(class_strings)].copy()
filtered_class_data_test.head()
```

```
# Add new columns that are required for YOLO format

filtered_class_data_train['classNumber'] = ''
filtered_class_data_train['center x'] = ''
filtered_class_data_train['center y'] = ''
filtered_class_data_train['width'] = ''
filtered_class_data_train['height'] = ''

filtered_class_data_test['classNumber'] = ''
filtered_class_data_test['center x'] = ''
filtered_class_data_test['center y'] = ''
filtered_class_data_test['width'] = ''
filtered_class_data_test['height'] = ''
```

```
# Iterate through all the class strings and assign a class number according to the order they are appearing in the list
# Here, 0 for Ambulance, 1 for Car and 2 for Taxi
for i in range(len(class_strings)):
  # Store the result of this operation into classNumber
  filtered_class_data_train.loc[filtered_class_data_train['LabelName'] == class_strings[i], 'classNumber'] = i
  filtered_class_data_test.loc[filtered_class_data_test['LabelName'] == class_strings[i], 'classNumber'] = i

# Calculate center x and center y values
filtered_class_data_train['center x'] = (filtered_class_data_train['XMax'] + filtered_class_data_train['XMin']) / 2
filtered_class_data_train['center y'] = (filtered_class_data_train['YMax'] + filtered_class_data_train['YMin']) / 2
filtered_class_data_test['center x'] = (filtered_class_data_test['XMax'] + filtered_class_data_test['XMin']) / 2
filtered_class_data_test['center y'] = (filtered_class_data_test['YMax'] + filtered_class_data_test['YMin']) / 2

# Calculate width and height valeus
filtered_class_data_train['width'] = filtered_class_data_train['XMax'] - filtered_class_data_train['XMin']
filtered_class_data_train['height'] = filtered_class_data_train['YMax'] - filtered_class_data_train['YMin']
filtered_class_data_test['width'] = filtered_class_data_test['XMax'] - filtered_class_data_test['XMin']
filtered_class_data_test['height'] = filtered_class_data_test['YMax'] - filtered_class_data_test['YMin']

# Generate the dataframe with YOLO required values
YOLO_values_train = filtered_class_data_train.loc[:, ['ImageID', 'classNumber', 'center x', 'center y', 'width', 'height']].copy()
YOLO_values_train.head()
```

```
# Generate the dataframe with YOLO required values
YOLO_values_test = filtered_class_data_test.loc[:, ['ImageID', 'classNumber', 'center x', 'center y', 'width', 'height']].copy()
YOLO_values_test.head()
```

```
# Change the current directory to the one where we have images
image_path = '/content/drive/My Drive/Colab Notebooks/DatasetnAnnotations/Dataset/train/Ambulance_Car_Taxi/'
os.chdir(image_path)

# Go through all the image files in the directory
# Fullstop in os.walk('.') means the current directory
for current_dir, dirs, files in os.walk('.'):
  # Iterating through all the files
  for f in files:
    # Check if the file extension ends with '.jpg'
    if f.endswith('.jpg'):
      # Extract only the title of the image file so that corresponding txt file can be generated with same name as per YOLO format
      image_title = f[:-4]
      # Get matched data
      YOLO_file = YOLO_values_train.loc[YOLO_values_train['ImageID'] == image_title]

      # Create a copy of the above dataframe
      df = YOLO_file.loc[:, ['classNumber', 'center x', 'center y', 'width', 'height']].copy()

      save_path = image_path + '/' + image_title + '.txt'

      # Generate a text file containing required data in YOLO format
      df.to_csv(save_path, header=False, index=False, sep=' ')
```
